```python
import pandas as pd
## package for apriori algo ##
## %pip install mlxtend ##
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
groceries = []
```

```python
## importing string data with 'open' function ##
with open (r'C:\Users\hp\Documents\forbidden\Hidden_Databases\groceries.csv') as f:
    groceries = f.read()
```

```python
groceries
```

```python
groceries = groceries.split("\n")
groceries
```

```python
groceries_list = []
```

```python
for i in groceries:
    groceries_list.append(i.split(','))
```

```python
groceries_list
```

```python
all_gro_list = [i for item in groceries_list for i in item]
all_gro_list
```

```python
## importing counter to count frequencies ##
from collections import Counter 
```

```python
## Coounting frequencies of each item in all_gro_list ##
item_freq = Counter(all_gro_list)
item_freq
```

```python
## Sorting item frequencies according to i[1] i.e. frequency position ##
item_freq = sorted(item_freq.items(), key = lambda x:x[1])
item_freq
```

```python
## Storing items in separate variables ##
frequencies = list(reversed([i[1] for i in item_freq]))
frequencies
```

```python
## Storing items in separate variables ##
items = list(reversed([i[0] for i in item_freq]))
items 
```

```python
## import package for plotting ##
import matplotlib.pyplot as plt
```

```python
## Bar plot (Bar Graph) ##
plt.bar(height = frequencies[1:11], x = list(range(1, 11)), color = 'hotpink')
plt.xticks(list(range(1, 11), ), items[1:11], rotation=90)
plt.xlabel("items")
plt.ylabel("Count")
plt.show()
```

```python
## ## Creating dataframe for the transaction data ##
groceries_series = pd.DataFrame(pd.Series(groceries_list))
groceries_series
```

```python
## removing the last null transaction ##
groceries_series = groceries_series.iloc[:9835, :]
```

```python
## giving name to the column as transaction ##
groceries_series.columns = ["transactions"]
groceries_series
```

```python
## creating a dummy columns for the each item in each transactions ##
## Using column names as item name ##
X = groceries_series['transactions'].str.join(sep = '*').str.get_dummies(sep = '*')
X
```

```python
## The support is computed as the fraction transactions where item occur/total_transaction ##
freq_itemsets = apriori(X, min_support = 0.0075, max_len = 4, use_colnames = True)
```

```python
freq_itemsets
## One will get a frozen datasets ##
```

```python
## Most frequent item sets based on support ##
freq_itemsets.sort_values('support',ascending=False, inplace=True)
freq_itemsets
```

```python
## Bar plot freq itemset & support ##
plt.bar(x = list(range(0, 11)), height = freq_itemsets.support[0:11], color = 'darkred')
plt.xticks(list(range(1, 11), ), freq_itemsets.itemsets[1:11], rotation=91)
plt.xlabel("item-sets")
plt.ylabel("support")
plt.show()
```

```python
## To find the association rules on frequent itemset ##
rules = association_rules(freq_itemsets, metric = "lift", min_threshold = 1)
## Top 20 association rules ##
rules
```

```python
## sorting rules according to 'lift' ##
Rules = rules.sort_values('lift', ascending = False)
## sorting rules according to 'lift' ##
Rules
```
